```python
pokedex=pd.read_csv(r"pokemon.csv")
pokedex
pokedex.describe()
```

```python
pokedex.head()
```

```python
pokedex.isnull().sum()
```

```python
stats=['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']
pokedex.info()
```

```python
generation_stats = pokedex.groupby('Generation')[stats].mean()
generation_stats

poke_type=pokedex.groupby("Type 1")[stats].median()
poke_type
```

```python
nonull=pokedex.dropna()
nonull
```

```python
plt.figure(figsize=(12, 6))
generation_stats.plot(marker='o')
plt.title('Generation-wise Evolution of Stats')
plt.xlabel('Generation')
plt.ylabel('Mean Stat Value')
plt.grid(True)
plt.legend(title='Stat Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
pokedex[stats].hist(figsize=(10, 10), bins=20)  
plt.suptitle('Distribution of Base Stats')  
plt.show()
```

```python


# Distribution of Type 1  
plt.figure(figsize=(12, 6))  
sns.countplot(x='Type 1', data=pokedex, order=pokedex['Type 1'].value_counts().index)  
plt.title('Distribution of Pokémon Type 1')  
plt.xticks(rotation=45)  
plt.show()  
#distribution of zType 2
plt.figure(figsize=(12, 6))  
sns.countplot(x='Type 2', data=pokedex, order=pokedex['Type 2'].dropna().value_counts().index)  
plt.title('Distribution of Pokémon Type 2')  
plt.xticks(rotation=45)  
plt.show()
```

```python
type_stats = pokedex['Type 1'].value_counts().reset_index()
type_stats.columns = ['Type', 'Frequency']
plt.figure(figsize=(12, 6))
sns.barplot(x='Type', y='Frequency', data=type_stats)
plt.title('Type Popularity')
plt.xlabel('Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='Type 1', y='Total', data=pokedex)
plt.title('Stat Distributions by Type')
plt.xlabel('Type')
plt.ylabel('Total Stat Value')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))  

legendary = pokedex[pokedex['Legendary'] == True]  
non_legendary = pokedex[pokedex['Legendary'] == False]  

plt.scatter(  
    x=legendary['Attack'],  
    y=legendary['Defense'],  
    marker='s',  # Square marker for legendary Pokémon  
    color='gold',  
    s=100,  # Size of the markers  
    label='Legendary Pokémon'  
)  

plt.scatter(  
    x=non_legendary['Attack'],  
    y=non_legendary['Defense'],  
    marker='o',  
    color='lightblue',  
    s=100,   
    label='Non-Legendary Pokémon'  
)  

plt.title('Pokémon Attack vs. Defense by Legendary Status')  
plt.xlabel('Attack')  
plt.ylabel('Defense')  
plt.legend(title='Legendary Status', loc='upper left')  

plt.grid()  
plt.show()  
```
