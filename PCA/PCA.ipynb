```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Load your dataset (replace 'your_dataset.csv' with your file)
data = pd.read_csv("/content/american_bankruptcy.csv").head(1000)
print(data.head())

# Extract numeric columns for PCA (excluding non-numeric like 'company_name' and 'status_label')
numeric_cols = data.select_dtypes(include=[np.number])


# Standardize the data (manual implementation)
mean_vals = numeric_cols.mean()
std_vals = numeric_cols.std()
scaled_data = (numeric_cols - mean_vals) / std_vals
scaled_data = scaled_data.values


# Perform PCA without using external libraries
def compute_pca(data):
    # Calculate the covariance matrix
    covariance_matrix = np.cov(data, rowvar=False)

    # Eigen decomposition
    eigen_values, eigen_vectors = np.linalg.eigh(covariance_matrix)

    # Sort eigenvalues and eigenvectors in descending order
    sorted_indices = np.argsort(eigen_values)[::-1]
    eigen_values = eigen_values[sorted_indices]
    eigen_vectors = eigen_vectors[:, sorted_indices]

    # Project data onto principal components
    principal_components = np.dot(data, eigen_vectors)
    return eigen_values, eigen_vectors, principal_components

# Compute PCA
eigen_values, eigen_vectors, pca_data = compute_pca(scaled_data)

# Variance explained by each component
explained_variance_ratio = eigen_values / np.sum(eigen_values)

# Scree Plot
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o', linestyle='--', color='b')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.grid()
plt.show()

# Biplot Function
def biplot(score, coeff, labels=None):
    xs = score[:, 0]
    ys = score[:, 1]
    plt.figure(figsize=(10, 7))
    plt.scatter(xs, ys, alpha=0.5, c='blue', label='Samples')
    for i in range(coeff.shape[0]):
        plt.arrow(0, 0, coeff[i, 0]*max(xs), coeff[i, 1]*max(ys),
                  color='red', alpha=0.7, head_width=0.02)
        if labels is None:
            plt.text(coeff[i, 0]*max(xs)*1.2, coeff[i, 1]*max(ys)*1.2, f"Var{i+1}", color='green')
        else:
            plt.text(coeff[i, 0]*max(xs)*1.2, coeff[i, 1]*max(ys)*1.2, labels[i], color='green')
    plt.xlabel('PC1')
    plt.ylabel('PC2')
    plt.grid()
    plt.legend()
    plt.title('Biplot')
    plt.show()

# Generate the biplot
biplot(pca_data, np.transpose(eigen_vectors), labels=numeric_cols.columns)

# Swiss Roll Plot (2D Visualization using the first two PCA components)
plt.figure(figsize=(10, 7))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=scaled_data[:, 0], cmap='viridis', alpha=0.7)
plt.title('Swiss Roll Plot (2D Visualization using PCA)')
plt.colorbar(label='Feature Intensity')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid()
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D

# Load dataset (use your actual dataset path)
df = pd.read_csv("/content/american_bankruptcy.csv").head(5000)

# Basic EDA
print("First few rows of the dataset:")
print(df.head())  # Show first few rows
print("\nData Information:")
print(df.info())  # Check for missing values and data types

# Check for any missing values
missing_data = df.isnull().sum()
print("\nMissing data:")
print(missing_data)

# Drop 'company_name', 'status_label', and 'year' columns as they are categorical and not needed for PCA
df_cleaned = df.drop(columns=['company_name', 'status_label', 'year'])

# Standardize the data (important for PCA)
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_cleaned)

# Apply PCA with custom parameters
n_components = 2  # Choose the number of components for dimensionality reduction
pca = PCA(n_components=n_components, whiten=True, svd_solver='auto')

# Fit and transform the data
df_pca = pca.fit_transform(df_scaled)

# Explained variance ratio (for the first 2 components in this case)
print(f"\nExplained variance ratio for the first {n_components} components: {pca.explained_variance_ratio_}")

# Plot Scree plot (Explained variance vs Components)
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o')
plt.title('Scree Plot')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.grid()
plt.show()

# Plot cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.show()

# Biplot for PCA (First two principal components)
def biplot(score, coeff, labels=None):
    xs = score[:, 0]
    ys = score[:, 1]
    plt.figure(figsize=(8, 6))
    plt.scatter(xs, ys, alpha=0.5)

    for i in range(len(xs)):
        if labels is None:
            plt.text(xs[i], ys[i], str(i), color='red', alpha=0.5)
        else:
            plt.text(xs[i], ys[i], labels[i], color='red', alpha=0.5)

    for i in range(len(coeff)):
        plt.arrow(0, 0, coeff[i, 0], coeff[i, 1], color='blue', alpha=0.5)
        plt.text(coeff[i, 0] * 1.1, coeff[i, 1] * 1.1, f'X{i+1}', color='blue', ha='center', va='center')

    plt.title('PCA Biplot')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.grid()
    plt.show()

# Biplot using the first two principal components
biplot(df_pca, pca.components_[:2, :])

# Swiss Roll Plot
# For this, we will first create a Swiss Roll-like dataset using sklearn (as a demonstration)
from sklearn import datasets
from sklearn.manifold import Isomap

# Generating Swiss roll dataset
swiss_roll_data, _ = datasets.make_swiss_roll(n_samples=500)

# Isomap to reduce to 2D for visualization
isomap = Isomap(n_components=2)
swiss_roll_reduced = isomap.fit_transform(swiss_roll_data)

plt.figure(figsize=(10, 6))
plt.scatter(swiss_roll_reduced[:, 0], swiss_roll_reduced[:, 1], c=swiss_roll_data[:, 0], cmap=plt.cm.Spectral)
plt.colorbar()
plt.title('Swiss Roll Plot')
plt.show()

# If you'd like to use your dataset for the Swiss Roll plot, it could be reduced to two dimensions using t-SNE or UMAP instead
```
